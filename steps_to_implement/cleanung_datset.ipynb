```python
# percentage foe pie chart (fires and not fires)
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# ploting piechart
classlabels = ["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of classes")
plt.show()
```

```python
## Monthly Fire analysis
df_reg1=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month',hue='Classes',data=df_reg1)
plt.xlabel("Month",weight='bold')
plt.ylabel("No Of fires",weight='bold')
plt.title("Fire in Sisi region")
```

It is observed that in Sisi region, during the August period there was maximum fire and during July region region did't experianced much fire
